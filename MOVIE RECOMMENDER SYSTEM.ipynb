```python
#converting movies['genres'] of json to useful string list

l =[]
for i in movies['genres']:
  json_array = i
  data = json.loads(json_array)
  formatted_genres = [genre['name'].strip() for genre in data]
  l.append(formatted_genres)

movies['genres']=l


m =[]
for i in movies['keywords']:
  json_array = i
  data = json.loads(json_array)
  formatted_data = [keywords['name'].strip() for keywords in data]
  m.append(formatted_data)
movies['keywords']=m
```

```python
#for sorting cast data and taking the first 3 leads or first 3 actors
n =[]
for i in movies['cast']:
  json_array = i
  data = json.loads(json_array)
  formatted_data = [keywords['name'].strip() for keywords in data]
  n.append(formatted_data)
o =[]
for i in n:
  j = i[:3]
  o.append(j)
n.clear()
movies['cast']=o
```

```python
#converting overview string to list
s = []
for i in movies['overview']:
    i = i.split()
    s.append(i)
movies['overview'] = s
```

```python
movies.head()
```

```python
#for removing all the blank space between every word such that it is easy for machine to differenciat between two words


movies['cast'] = [[name.replace(" ","").split(',') for name in sublist]for sublist in movies['cast']]
```

```python
movies['cast'].head()
```

```python
# now we have to remove sublist from every list file

p = []
for i in movies['cast']:
  flattened_list = [item for sublist in i for item in sublist]
  p.append(flattened_list)
movies['cast'] = p
```

```python
movies['cast']
```

```python
# now repeting it for cast, genres, keywords
movies['keywords'] = [[name.replace(" ","").split(',') for name in sublist]for sublist in movies['keywords']]
movies['genres'] = [[name.replace(" ","").split(',') for name in sublist]for sublist in movies['genres']]
movies['crew'] = [[name.replace(" ","").split(',') for name in sublist]for sublist in movies['crew']]

p = []
for i in movies['keywords']:
  flattened_list = [item for sublist in i for item in sublist]
  p.append(flattened_list)
movies['keywords'] = p

p = []
for i in movies['genres']:
  flattened_list = [item for sublist in i for item in sublist]
  p.append(flattened_list)
movies['genres'] = p

p = []
for i in movies['crew']:
  flattened_list = [item for sublist in i for item in sublist]
  p.append(flattened_list)
movies['crew'] = p
```

```python
movies.head()
```

```python
#now let's merge overview , genres , keywords , cast , crew as "tag" named column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies.drop(['overview' , 'genres' , 'keywords' , 'cast' , 'crew'],axis = 1,inplace =True)
```

```python
#connecting all lists in one single string in tags and lowering all the sentence

list1 =[]
for i in movies['tags']:
  i = " ".join(i)
  i = i.lower()
  list1.append(i)

movies['tags'] = list1
```

```python
#now lets do vectorization with top 5000 words and also remove stopwords in english lang

cv = cv(max_features=5000,stop_words='english')
```

```python
#as cv will return skipi matrix and we required numpy array as output so .toarray()
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
# here most of the vector are zero because we merged all tags file from movie 1 to end and made a bag of words with in from there we are picking top 5000 words and
# comparing it with all individual tags thats why repeatability most of most of the word in single tag is zero but some words are repeated
# vectors
```

```python
feature_list = cv.get_feature_names_out()
array = feature_list

# Set print options to display the full array
np.set_printoptions(threshold=np.inf)

# Print the entire array
print(array)
```

```python
# for applying stamming on this list to reduce words by making all [loving,love,loved] as love
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stemmer(text):
  y = []
  for i in text.split():
      y.append(ps.stem(i))

  #now converting this list in string again
  return " ".join(y)
```

```python
ps.stem('loving')
```

```python
stemmer(movies['tags'][0])
```

```python
movies['tags']=movies['tags'].apply(stemmer)
```

```python
#now lets apply all vectorization on stammered data again
from sklearn.feature_extraction.text import CountVectorizer as cv
cv = cv(max_features=5000,stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()
feature_list = cv.get_feature_names_out()
array = feature_list

# Set print options to display the full array
np.set_printoptions(threshold=np.inf)

# Print the entire array
print(array)
```

```python
#here now for similarity purposes, we will calculate the distance between each movie with every movie but the thing is Euclidian distance will not be that accurate for high-dimension data
# So we will calculate cosine distance which is distance in angle between two vectors such that if 2 vector has a distance of 180 degrees than they are movies of opposite genres

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
#for a recommandation first lets fatch the index of the movie to sort this similarity data


movie_index = movies[movies['title'] == "Batman Begins"].index[0]
movie_index

# now after fatching movie index we have to measure distance from other movies for that we have to pass this data from that perticular similarity matrix
distance = similarity[movie_index]
sorted(similarity[movie_index], reverse=True)[1:6]


```

```python
#but the thing is by sorting this we will lose index of the movie to fatch back the name of the movie
#so we can use enumerate fn to add index in any list by converting list into tuple
tuple_with_index = list(enumerate(similarity[movie_index]))

#and now to sort on movie_index bases we will write
rec_movies = sorted(tuple_with_index,reverse=True,key=lambda x:x[1])[1:6]
```

```python
for i in range(len(rec_movies)):
  print(movies.iloc[rec_movies[i][0]].title)
```

```python
# for recommanding a movie lets create a fn
def recommend(movie):
  movie_index = movies[movies['title'] == movie].index[0]
  distance = similarity[movie_index]
  tuple_with_index = list(enumerate(similarity[movie_index]))
  movies_list = sorted(tuple_with_index,reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(movies.iloc[i[0]].title)
```

```python
recommend('Jab Tak Hai Jaan')
```
